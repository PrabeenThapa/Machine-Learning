### Handling Duplicates (Find And Remove)

Handling Duplicates in Python Pandas means to find and remove duplicates from rows in a DataFrame or Series, use the:

<ol>
<li><b>duplicated()</b>: Find duplicates</li>
<li><b>drop_duplicates()</b>: Remove duplicates</li>
</ol>

#### 1. Find Duplicates

To find duplicats from rows in a Pandas DataFrame or Series, usee the <b>duplicated</b> method. It resturns a Series with True and False values i.e for duplicate rows true is returned.

```python
import pandas as pd

#reading through CSV

df = pd.read_csv("/home/prabin/Desktop/python/Pandas/_pandas/Earthquake_Perception_Dataset.csv")

df
```

```python
#find duplicates

resDuplicate = df.duplicated()

print("\n",resDuplicate)
```

#### 2. Remove Duplicates

```python
#removing duplicates using drop duplicates

df.drop_duplicates()

df
```
