### Pandas Clean The Data

To clean the data in python, we have some built-in functions. Cleaning  the data in Pandas means working on the incorrect data to fix it. This incorrect data can be empty data, null, duplicate data etc. 

```python
import pandas as pd

#dataframe

df = pd.read_csv("/home/prabin/Desktop/python/Pandas/_pandas/demo.csv")

df
```

#### Buit-In functions:
<ol>
<li><b>isnull()</b>: Find the NULL values and replace them with True.</li>
<li><b>notnull()</b>: Find the NOT NULL values and replace them with True.</li>
<li><b>df.dropna()</b>: Drop rows with NULL values.</li>
<li><b>df.fillna(x)</b>: Replace NULL values with a specific value.</li>
</ol>

```python
# isnull()

df.isnull()
```

```python
# notnull()

df.notnull()
```

```python
# dropna()

resDF = df.dropna()

resDF
```

```python
# fillna()

resDF = df.fillna(404)

resDF
```
