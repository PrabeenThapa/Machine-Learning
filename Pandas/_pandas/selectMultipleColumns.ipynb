### Select Multiple Columns

In Python, easily select multiple columns in a Pandas DataFrame. Select more than one column without using built-in functions. More than two columns can also be selected in a range.

#### 1. Select two columns

```python
import pandas as pd

#dataframe

df = pd.read_csv("/home/prabin/Desktop/python/Pandas/_pandas/Earthquake_Perception_Dataset.csv")
df = pd.DataFrame(df)
df
```

```python
# selecting two columns

df[['age', 'residence']]
```

#### 2. Select Multiple Columns In A Range

In a Pandas Dataframe, to select more than one column in a range, mention the index numbers in a range separated by a colon.

```python
#selecting multiple columns

df[df.columns[1:3]]
```
