### Delete Rows/Column

The <b>drop()</b> method is used to delete rows and column in Pandas DataFrame. 

#### 1. Delete Columns 

The columns are dropped using the column names. The <b>axis</b> is set to 1 since we want to drop a column. The <b>columns</b> axis can also be used for the <b>drop()</b> method to remove the specified column.

```python
import pandas as pd

df = pd.read_csv("/home/prabin/Desktop/python/Pandas/_pandas/Earthquake_Perception_Dataset.csv")

print("\nOriginal DataFrame:\n\n",df)
```

```python
#deleting columns
resDF = df.drop(labels="age", axis='columns')
print("After deleting age column: \n", resDF)
```

#### 2. Delete rows

The rows are dropped using the rows names. The <b>axis</b> is set to 0 since we want to drop a rows. The <b>rows</b> axis i.e <b>index</b> can also be used for the <b>drop()</b> method to remove the specified rows.

```python
resDF = df.drop(labels=0, axis='index')

print("\nAfter removing first rows:\n\n", resDF)
```
