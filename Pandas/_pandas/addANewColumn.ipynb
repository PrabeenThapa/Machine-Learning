### Add A New Column

We can insert a column to an already created DataFrame by following two method:

#### 1. Add a new column using the <b>insert</b> method(set th location)

```python
import pandas as pd

df = pd.DataFrame(pd.read_csv("/home/prabin/Desktop/python/Pandas/_pandas/Earthquake_Perception_Dataset.csv"))
df
```

```python
df.insert(loc=2, column="added",value=[19,20,4,25])#upto 472 term is required

df
```

#### 2. Assign a new column using the <b>assign</b> method (place at the end)

```python
df.assign(Roll = [2,1,4,3,5,7]) #upto 472 terms

df
```
